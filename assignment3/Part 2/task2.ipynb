```python
import matplotlib.pyplot as plt
from PIL import Image

def show_images(num_list):
    total_num = 938 * 200
    num_rows = (len(num_list) + 4) // 5
    fig, axes = plt.subplots(num_rows, 5, figsize=(20, 4*num_rows))
    for i, num in enumerate(num_list):
        path = './task2/{}.png'.format(num)
        img = Image.open(path)
        ax = axes[i // 5, i % 5] if num_rows > 1 else axes[i % 5]
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
    
```

```python
start_list = [0, 1500, 3000, 4500, 6000]
print(start_list)
show_images(start_list)
```

```python
halfway_list = [90000, 95000, 100000, 105000, 110000]
print(halfway_list)
show_images(halfway_list)
```

```python
end_list = [170000, 175000, 180000, 185000, 187500]
print(end_list)
show_images(end_list)
```
